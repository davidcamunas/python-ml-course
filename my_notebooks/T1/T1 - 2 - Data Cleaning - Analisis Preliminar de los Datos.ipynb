## Valores que Faltan (Missing values)

#### Si es NULL

```python
pd.isnull(titanic_df["body"])
```

#### Si no es NULL

```python
## Para convertir a Array, y sumarlos en este caso los Trues
pd.notnull(titanic_df["body"]).values.sum()
```

Los valores que faltan en un datasheet pueden venir por dos razones:
* ExtracciÃ³n de los datos
* Recoleccion de los datos


#### Borrado de valores que faltan

* axis = 1 borra columnas
* axis = 0 borra filas

```python
# los que todas las columnas sean Nan
titanic_df.dropna(axis=0, how="all")
```

```python
# Todo el cual alguna columna contenga un Nan
titanic_df2.dropna(axis=0, how="any")
```

#### Computo de los valores que faltan

```python
titanic_df2 = titanic_df
# Todo valor que sea Nan se sustituye con 0
titanic_df2.fillna(0)
```

```python
titanic_df3 = titanic_df
titanic_df3.fillna("Desconocido")
```

```python
titanic_df4 = titanic_df
```

```python
# Los Nan de esta columna se reemplazaran por 0
titanic_df4["body"] = titanic_df4["body"].fillna(0)
titanic_df4["home.dest"] = titanic_df4["home.dest"].fillna("Desconocido")
titanic_df4.head()
```

```python
titanic_df4["home.dest"].fillna("Desconocido")
```

```python
pd.isnull(titanic_df4["age"]).values.sum()
```

```python
# Sustituir los valores nulos de la columna aÃ±o por la media
titanic_df4["age"].fillna(titanic_df4["age"].mean())
```

```python
# En este caso pone el valor de la fila anterior al Nan
titanic_df4["age"].fillna(method="ffill")
```

```python
# Lo mismo pero reemplazdo por el primero hacia atras en este caso
titanic_df4["age"].fillna(method="backfill")
```

## Variables Dummy

```python
titanic_df4["sex"]
```

```python
# Un dato dymmy es la creaciÃ³n de unas nuevas filas a partir de una. Creando asÃ­ N filas por cada valor existente para esa columna. Es decir, en este caso la columna sex puede ser female o male. 
# Por lo tanto, se puede generar una columna en el cual se defina de forma booleana si la fila tiene el valor male o female.

dummy_sex = pd.get_dummies(titanic_df4["sex"], prefix="sex")
dummy_sex.head(8)
```
